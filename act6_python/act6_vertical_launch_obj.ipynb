### 1. Consider an object thrown up into the air on a vertical trajectory. The vertical position at each time can be described by: $$ y = v_0 t - \frac{1}{2} g t^2 $$ where $ v_0 $ is the initial velocity, and $ g $ is the acceleration due to gravity.

 

A. Find the maximum height between $t = 0$ and $2s$ for $v_0 = 5 ~m/s$

```python
import numpy as np

# Constants
g = 9.81  # Acceleration due to gravity (m/s^2)
v0 = 5.0  # Initial velocity (m/s)
t_max = 2.0  # Maximum time (s)
num_steps = 2000  # Number of time steps

# Time array
t = np.linspace(0, t_max, num_steps)

# Vertical position function
def vertical_position(t, v0, g):
    return v0 * t - 0.5 * g * t**2

y = vertical_position(t, v0, g)
max_height = np.max(y)
print(f"Maximum height (NumPy): {max_height:.2f} m")
```

B. Write a for loop that iterates over 2000 time-steps between 0 and 2 and find the time step with the largest y value by comparing it to y from the previous step.

```python
for i in range(1, len(t)):
    if max_height < y[i]:
        max_height = y[i]
        max_time = t[i]
print(f"Maximum height (Loop): {max_height:.2f} m at time {max_time:.2f} s")
```

C. Alternatively, you can write a while loop that compares the value of $y_{i+1}$ and $y_i$.

```python
i = 0
while True:
    if y[i + 1] > y[i]:
        max_height = y[i + 1]
        max_time = t[i + 1]
        i += 1
    else:
        break
print(f"Maximum height (While Loop): {max_height:.2f} m at time {max_time:.2f} s")
```

D. Create a plot, that shows the objectâ€™s trajectory as a black curve and the maximum value highlighted by a red star.

```python
import matplotlib.pyplot as plt
 
fig, ax = plt.subplots()
ax.plot(t, y, 'k-', label='Trajectory')
ax.plot(max_time, max_height, 'r*', markersize=12, label='Maximum Height')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Height (m)')
ax.set_title('Vertical Launch Trajectory')
ax.legend()
ax.grid(linewidth=0.5, linestyle='--')
plt.show()
```

### 2. Determine the derivative and plot the derivative over time in a second subplot (subplot(211)). Explain your observations!

```python
# calculate derivative with numpy

dy_dt = np.gradient(y, t)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(t, y, 'k-', label='Trajectory')
ax1.plot(max_time, max_height, 'r*', markersize=12, label='Maximum Height')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Height (m)')
ax1.set_title('Vertical Launch Trajectory')
ax1.legend()
ax1.grid(linewidth=0.5, linestyle='--')
ax2.plot(t, dy_dt, 'b-', label='Derivative (Velocity)')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Velocity (m/s)')
ax2.set_title('Velocity over Time')
ax2.legend()
ax2.grid(linewidth=0.5, linestyle='--')
plt.tight_layout()
plt.show()
```

<p><b>Explanation:</b></p>

From the plot above, we can observe that the derivative of the vertical position (which represents the vertical velocity) decreases linearly over time. This is consistent with the physics of an object under the influence of gravity, where the velocity decreases at a constant rate due to the acceleration caused by gravity and is reflected in the  negative slope of the velocity-time graph. The point where the velocity reaches zero corresponds to the maximum height of the object, after which the velocity becomes negative as the object begins to descend back.
